```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize lists to store scraped data
data_list = []

# Determine the total number of pages
total_pages = 60 # Set it according to the number of pages you want to scrape

# Function to scrape data from a single page
for page_num in range(1, total_pages + 1):
    
    # Base URL of the website to scrape
    url = 'https://startupmahakumbh.assocham.org/exhibitor_directory/exhi_list_pub.php?event_name=sm&event_year=2024&page='+str(page_num)+'&filter=All'
    
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        details_divs = soup.find_all('div', class_='details')

        for div in details_divs:
            # Find all p elements within the div
            p_tags = div.find_all('p')

            # Extract text from p tags and append to data_list
#             data = [p.text for p in p_tags]
            data = [p.text.split(':', 1)[1].strip() for p in p_tags]  # Extract text after the colon

            data_list.append(data)        
        
    else:
        print(f'Failed to retrieve {url}')

# Convert data_list into a Pandas DataFrame
df = pd.DataFrame(data_list, columns=['Company Name','Address','Contact Person','Designation','Phone','Profile'])  
df=df.drop(columns=['Profile'])

# Display the DataFrame
df
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfFileReader(file)
        num_pages = pdf_reader.numPages
        for page_number in range(num_pages):
            page = pdf_reader.getPage(page_number)
            text += page.extractText()
    return text

# Path to your PDF file
pdf_path = r"C:\Users\nimbu\Downloads\Exhibitor_Directory.pdf"

# Extract text from the PDF file
pdf_text = extract_text_from_pdf(pdf_path)

print(pdf_text)
```

```python
import pandas as pd
import re

# Extracting data using regular expressions
name_matches = re.compile(r"Name of the Exhibitor\s*:\s*(.+)").finditer(pdf_text)
address_matches = re.compile(r"Address\s*:\s*(.+)").finditer(pdf_text)
contact_person_matches = re.compile(r"Contact Person\s*:\s*(.+)").finditer(pdf_text)
designation_matches = re.compile(r"Designation\s*:\s*(.+)").finditer(pdf_text)
mobile_matches = re.compile(r"Mobile\s*:\s*(.+)").finditer(pdf_text)
email_matches = re.compile(r"Email\s*:\s*(.+)").finditer(pdf_text)
website_matches = re.compile(r"Website\s*:\s*(.+)").finditer(pdf_text)
sector_matches = re.compile(r"Sector\s*:\s*(.+)").finditer(pdf_text)

# Creating lists to store extracted data
names = [match.group(1).strip() if match else None for match in name_matches]
addresses = [match.group(1).strip() if match else None for match in address_matches]
contact_persons = [match.group(1).strip() if match else None for match in contact_person_matches]
designations = [match.group(1).strip() if match else None for match in designation_matches]
mobiles = [match.group(1).strip() if match else None for match in mobile_matches]
emails = [match.group(1).strip() if match else None for match in email_matches]
websites = [match.group(1).strip() if match else None for match in website_matches]
sectors = [match.group(1).strip() if match else None for match in sector_matches]

# Creating DataFrame
df = pd.DataFrame({
    "Company": names,
    "Address": addresses,
    "Contact Person": contact_persons,
    "Designation": designations,
    "Mobile": mobiles,
    "Email": emails,
    "Website": websites,
    "Sector": sectors
})

# Displaying the DataFrame
# print(df)
df
```
